```python
# 5. Convert datatype of column published_at from object to pandas datetime.
df['published_at']=pd.to_datetime(df['published_at'])
df.info()
```

```python
# 6. Create a new column as 'published_month' using the column published_at (display the months only)
df['published_month'] = df['published_at'].dt.month
df
```

```python
# 7. Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb'and so on.....
df['published_month']=df['published_at'].apply(lambda x:x.month_name())
df
```

```python
# 13. Does the number of views have any effect on how many people disliked the video? Support your answer with a metric and a plot.
# plot a scatter plot between views and dislikes
plt.scatter(df['view_count'], df['dislikes'])
plt.xlabel('view_count')
plt.ylabel('dislikes')
plt.show()
# calculate the correlation coefficient between views and dislikes
df['view_count'].corr(df['dislikes'])
```

* A correlation coefficient of 0.6846 suggests a moderate positive correlation between the view_count and the number of dislikes. 
* It means that as the view count increases, the number of dislikes also tends to increase. 
* However, it is important to note that correlation coefficient value is very small, so we cannot conclude that views are causing dislikes or vice versa. 
* There may be other factors that are influencing both variables
